**Exercise 1 (10 pts)**. Avoding catastrophic cancellation.

The tail of the standard logistic distributon is given by $1 - F(t) = 1 - (1+e^{-t})^{-1}$.

- Define a function `f1` to calculate the tail probability of the logistic distribution using the formula given above
- Use `sympy` to find the exact value of the tail distribution (using the same symbolic formula) to 20 decimal digits
- Calculate the *relative error* of `f1` when $t = 25$ (The relative error is given by `abs(exact - approximate)/exact`)
- Rewrite the expression for the tail of the logistic distribution using simple algebra so that there is no risk of cancellation, and write a function `f2` using this formula. Calculate the *relative error* of `f2` when $t = 25$. 
- How much more accurate is `f2` compared with `f1` in terms of the relative error?

````python
# Your code here
def f1(t):
    return 1-(1+np.exp(-t))**(-1)

print "Tail probability calculated using f1(25) is:", f1(25)
 
t = symbols('t')
exact = 1-(1+exp(-t))**(-1)
exact20 = N(exact.subs(t,25),20)
print "Exact tail probability to 20 decimal points calculated using sympy(for t = 25) is:", exact20

errorf1 = abs((f1(25) - exact20)/exact20)
print "Relative error of f1 for t = 25 is:", errorf1

def f2(t):
    return np.exp(-t)/(1+np.exp(-t))

errorf2 = abs((f2(25) - exact20)/exact20)
print "Relative error of f2 for t = 25 is:", errorf2

print "f2 is", errorf1/errorf2, "times more accurate compared to f1."
````

**Exercise 2 (10 pts)**. Ill-conditioned linear problems.

You are given a $n \times p$ design matrix $X$ and a $p$-vector of observations $y$ and asked to find the coefficients $\beta$ that solve the linear equations $X \beta = y$.

- How could you code this formula using `np.linalg.solve` that does not require inverting a matrix? Write a function `f2` that takes arguments $X$ and $y$ and returns $\beta$ using this.
- Note that carefully designed algorithms *can* solve this ill-conditioned problem, which is why you should always use library functions for linear algebra rather than write your own.
```python

```
- What happens if you try to solve for $\beta$ using `f1` or `f2`? Remove the column of $X$ that is making the matrix singluar and find the $p-1$ vector $b$ using `f2`.
- Note that the solution differs from that given by `np.linalg.lstsq`? This arises because the relevant condition number for `f2` is actually for the matrix $X^TX$ while the condition number of `lstsq` is for the matrix $X$. Why is the condition so high even after removing the column that makes the matrix singular?

##### Answers:


- The equation that could solve the linear equations in terms of X and y is:
$\beta = (X^T X)^{-1} X^Ty$
- To code using np.linalg.solve without require inverting the matrix, we could use $X^TX$ and $X^Ty$ as arguments instead of $X,y$ for the np.linalg.solve function.

````python

X = np.load('x.npy')
y = np.load('y.npy')
beta = np.load('b.npy')

#function to solve the linear equations taking X and y as arguments
def f1(X,y):
    X_t = X.T
    X_t_X = np.dot(X_t,X)
    X_t_X_inv = linalg.inv(X_t_X)
    return np.dot(X_t_X_inv, np.dot(X_t, y))

# function to solve the equation that does not require inverting the matrix.
def f2(X,y):
    return np.linalg.solve(np.dot(X.T,X),np.dot(X.T,y))

print "Solution using np.linalg.lstsq:", np.linalg.lstsq(X, y)[0]

# find the column that's making the matrix singular and remove it
for i in range(0,6):
    if np.linalg.matrix_rank(np.delete(X,i,1))==5:
        X_1 = np.delete(X,i,1)
        break
print "Solution by removing a column that's making the matrix singular:", f2(X_1,y)
print "Condition numbers for X and X_1(matrix after deleted column) are:", np.linalg.cond(X),np.linalg.cond(X_1), "respectively"
print "X_1:",X_1
print "As we can see above, one of the columns in X_1 is orders of magnitude (14) bigger than others.\
This is the reason why we're getting such high conditionn number even after removing the column\
that makes the matrix singular."
````

**Exercise 3 (10 pts)**. Importance of using efficient algorihtms.

- Implement bubble sort
- Calculate its big $\mathcal{O}$ algorithmic complexity
- Time the performance of bubble sort on random uniform deviate vectors of sizes `range(100, 2000, 100)` using `time.time()` from the standard library
- Use `scipy.optimize.curve_fit` to fit an appropriate function to the collection of (size, execution time) data points. Extrapolate how long it would take to sort a random vector of size 1,000,000. Now time how long it takes for the system sort to sort a random vector of size 1,000,000.
- Plot the fits together with the data points uisng `matplotlib.pyplot` functions.

````python
np.random.seed(1000)

def bubble_sort(X):
    for i in range(len(X)-1,0,-1):
        for j in range(i):
            if X[j]>X[j+1]:
                x_j=X[j]
                X[j]=X[j+1]
                X[j+1]=x_j
    return X
X = [3,5,6,2,5,9,1,2,6,3,8,3,5,4,7]
print X,"is sorted into", bubble_sort(X), "by using the implementation of bubble sort \n"


print "The big O algorithmic complexity of the bubbble sort algorithm in O(n^2) on average.\
As we can see from the two nested loops, the we have n^2 order of computation which leads to O(n^2).\
In the best case when the list is already sorted, we have the complexity of O(n) \n"

runtimes=[]
#Calculate time needed to perfom bubble sort for different length lists.
for i in range(100, 2000, 100):
    list1=np.random.random(i) #generate random numbers
    start=time.time()
    bubble_sort(list1)
    end=time.time()
    runtimes.append(end-start)
print "Time required by bubble sort to sort lists of lengths from 100 to 2000:", runtimes, "\n"   
# Because of O(n^2) complexity, we're trying to fit it to a quadratic equation
#define a quadratic function
def quad(x,a,b,c):
    return a+b*x+c*(x**2) 

coef,s=curve_fit(quad,range(100, 2000, 100), runtimes) # estimate the coefficients using curve_fit

t1million = quad(1000000,coef[0],coef[1],coef[2])
print "Time required by bubble sort to sort a list of length 1 million is:", t1million, "\n" 

start=time.time()
sorted(np.random.uniform(1,100,1000000))
end=time.time()
print "Time required by system sort to sort the same list of same length:", end-start, "\n"


plt.plot(range(100, 2000, 100),runtimes,'bo')
xi = np.linspace(0,2000,100)
yi = quad(xi, coef[0], coef[1], coef[2])
plt.plot(xi, yi, 'r-')
````

**Exercise 4 (20 pts)**. One of the goals of the course it that you will be able to implement novel algorihtms from the literature. 

- Implement the mean-shift algorithm in 1D as described [here](http://homepages.inf.ed.ac.uk/rbf/CVonline/LOCAL_COPIES/TUZEL1/MeanShift.pdf). 
    - Use the following function signature
    ```python
    def mean_shift(xs, x, kernel, max_iters=100, tol=1e-6):
    ```
    - xs is the data set, x is the starting location, and kernel is a kernel function
    - tol is the difference in $||x||$ across iterations
- Use the following kernels with bandwidth $h$ (a default value of 1.0 will work fine)
    - Flat - return 1 if $||x|| < h$ and 0 otherwise
    - Gaussian 
    $$\frac{1}{\sqrt{2 \pi h}}e^{\frac{-||x||^2}{h^2}}$$
    - Note that $||x||$ is the norm of the data point being evaluated minus the current value of $x$
- Use both kernels to find all 3 modes of the data set in `x1d.npy`
- Modify the algorithm and/or kernels so that it now works in an arbitrary number of dimensions.
- Use both kernels to find all 3 modes of the data set in `x2d.npy`
- Plot the path of successive intermediate solutions of the mean-shift algorithm starting from `x0 = (-4, 10)` until it converges onto a mode in the 2D data for each kernel. Superimpose the path on top of a contour plot of the data density.

````python
# Your code here
from scipy.misc import derivative

# Mean shift function for 1D data
def mean_shift_1(xs, x, kernel, max_iters=100, tol=1e-6):
    """Calculate the local mode using mean shift clustering.
    Inputs:
        xs: 1D list
        x: starting point
        kernel: a kernel function to use
        max_iters: total iterations to run unless mode is found earlier
        tol: tolerance
    Returns:
        mode: local mode discovered
    """
    for j in range(max_iters):
        num = 0
        denom = 0
        for i in range(len(xs)):
            z = (np.linalg.norm(x-xs[i]))**2
            #num += xs[i] * (-derivative(kernel, z))
            num += xs[i] * kernel(z)
            #denom += (-derivative(kernel, z))
            denom += kernel(z)
        mean_shift = num/denom - x
        if abs(mean_shift) < tol:
            break
        else:
            x = x + mean_shift    
    mode = x
    return mode

#define gaussian and flat kernels
def gaussian(x, h=1):
    return 1/np.sqrt(2*np.pi)*np.exp(-x/h**2)
def flat(x, h=1):
    if np.sqrt(x)<h:
        return 1
    else:
        return 0
    
xs = np.load('x1d.npy')
modes_gaussian =[]
while len(modes_gaussian)<3:
    start = np.random.uniform(min(xs), max(xs))
    mode = round(mean_shift_1(xs,start, gaussian),4)
    if mode not in modes_gaussian:
        modes_gaussian.append(mode)
print "The three modes of 1D data set obtained using gaussian kernel are:", modes_gaussian  

#Using flat kernel did not return same mode within 4 decimal points regardless of starting point,
#so different set starting points were used.
modes_flat = [0,0,0]
modes_flat[0] = round(mean_shift_1(xs, min(xs), flat),4)
modes_flat[1] = round(mean_shift_1(xs, 0, flat),4)
modes_flat[2] = round(mean_shift_1(xs, max(xs), flat),4)
print "The three modes of 1D data set obtained using flat kernel are:", modes_flat
````

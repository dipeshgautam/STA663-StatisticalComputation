**Exercise 1 (10 points)**.  Calculating pairwise distance matrices.

Suppose we want to construct a distance matrix between the rows of a matrix. For example, given the matrix 

```python
M = np.array([[1,2,3],[4,5,6]])
```

the distance matrix using Euclidean distance as the measure would be
```python
[[ 0.000  1.414  2.828]
 [ 1.414  0.000  1.414]
 [ 2.828  1.414  0.000]] 
```
if $M$ was a collection of column vectors.

Write a function to calculate the pairwise-distance matrix given the matrix $M$ and some arbitrary distance function. Your functions should have the following signature:
```
def func_name(M, distance_func):
    pass
```

0. Write a distance function for the Euclidean, squared Euclidean and cosine measures.
1. Write the function using looping for M as a collection of row vectors.
2. Write the function using looping for M as a collection of column vectors.
3. Wrtie the function using broadcasting for M as a colleciton of row vectors.
4. Write the function using broadcasting for M as a colleciton of column vectors. 

For 3 and 4, try to avoid using transposition (but if you get stuck, there will be no penalty for using transpoition). Check that all four functions give the same result when applied to the given matrix $M$.

````python
def euclidean_dist(x,y):
    """Returns the Euclidean distance between two numpy vectors"""
    return np.sum((x -y)**2, -1)**0.5

def squared_euclidean_dist(x,y):
    """Returns the squared Euclidean distance between two numpy vectors"""
    return np.sum((x - y)**2, -1)

def cosine_dist(x,y):
    """Returns the cosine distance between two numpy vectors"""
    return 1 - np.sum(x*y,-1)/(np.sum(x**2,-1)**0.5*np.sum(y**2,-1)**0.5)

def row_vector_loop(M, distance_func):
    """Returns the pairwise distance matrix by looping and treating the given matrix as collection of row vectors.
        The provided distance function decides what distnace measure to calculate"""
    n = M.shape[0]
    dist = np.zeros((n,n))
    for i in range(n):
        for j in range(i):
            dist[i,j] = distance_func(M[i,:], M[j,:])
    dist = dist + dist.T - np.diag(np.diag(dist))
    return dist

def col_vector_loop(M, distance_func):
    """Returns the pairwise distance matrix by looping  and treating the given matrix as collection of column vectors.
        The provided distance function decides what distance measure to calculate."""
    n = M.shape[1]
    dist = np.zeros((n,n))
    for i in range(n):
        for j in range(i):
            dist[i,j] = distance_func(M[:,i], M[:,j]) 
    dist = dist + dist.T - np.diag(np.diag(dist))
    return dist

def row_vector_broadcast(M, distance_func):
    """Returns the pairwise distance matrix by using broadcast rules
        and treating the given matrix as collection of row vectors.
        The provided distance function decides what distance measure to calculate."""
    return distance_func(M[None,:], M[:,None])

def col_vector_broadcast(M, distance_func):
    """Returns the pairwise distance matrix by using broadcast rules
        and treating the given matrix as collection of column vectors.
        The provided distance function decides what distance measure to calculate."""    
    M=M.T
    return distance_func(M[None,:], M[:,None])
````

````python
M = np.array([[1,2,3],[4,5,6]])

print "Euclidean distance using loop for collection of row vectors: \n",row_vector_loop(M, euclidean_dist)
print "Euclidean distance using broadcasting for collection of row vectors: \n",row_vector_broadcast(M, euclidean_dist)
print "Squared euclidean distance using loop for collection of row vectors: \n", row_vector_loop(M, squared_euclidean_dist)
print "Squared euclidean distance using broadcasting for collection of row vectors: \n",row_vector_broadcast(M, squared_euclidean_dist)
print "Cosine distance using loop for collection of row vectors: \n", row_vector_loop(M, cosine_dist)
print "Cosine distance using broadcasting for collection of row vectors: \n",row_vector_broadcast(M, cosine_dist)


print "Euclidean distance using loop for collection of column vectors: \n",col_vector_loop(M, euclidean_dist)
print "Euclidean distance using broadcasting for collection of column vectors: \n",col_vector_broadcast(M, euclidean_dist)
print "Squared euclidean distance using loop for collection of column vectors: \n", col_vector_loop(M, squared_euclidean_dist)
print "Squared euclidean distance using broadcasting for collection of column vectors: \n",col_vector_broadcast(M, squared_euclidean_dist)
print "Cosine distance using loop for collection of column vectors: \n", col_vector_loop(M, cosine_dist)
print "Cosine distance using broadcasting for collection of column vectors: \n",col_vector_broadcast(M, cosine_dist)
````

**Exercise 2 (10 points)**. Write 3 functions to calculate the term frequency (tf), the inverse document frequency (idf) and the product (tf-idf). Each function should take a single argument `docs`, which is a dictionary of (key=identifier, value=dcoument text) pairs, and return an appropriately sized array. Convert '-' to ' ' (space), remove punctuation, convert text to lowercase and split on whitespace to generate a collection of terms from the dcoument text.

- tf = the number of occurrences of term $i$ in document $j$
- idf = $\log \frac{n}{1 + \text{df}_i}$ where $n$ is the total number of documents and $\text{df}_i$ is the number of documents in which term $i$ occurs.

Print the table of tf-idf values for the following document collection

```
s1 = "The quick brown fox"
s2 = "Brown fox jumps over the jumps jumps jumps"
s3 = "The the the lazy dog elephant."
s4 = "The the the the the dog peacock lion tiger elephant"

docs = {'s1': s1, 's2': s2, 's3': s3, 's4': s4}
```

Note: You can use either a numpy array or pandas dataframe to store the matrix. However, we suggest using a Pnadas dataframe since that will allow you to keep track of the row (term) and column (document) names in a single object. Of course, you could also maintain a numpy matrix, a list of terms, and a list of documents separately if you prefer.

````python
def tfs(docs):
    """Create a term freqeuncy dataframe from a dictionary of documents.
    We preprocess each document to strip punctuation and convert to lowercase.
    Terms are found by splitting on whitespace."""
    termFreq = {}
    for x in docs:
        words = filter(lambda c: c not in string.punctuation, docs[x].replace('-', ' ')).lower().split()
        #words = docs[x].translate(None, punctuation).lower().replace("-", " ").split()
        for word in words:
            if word not in termFreq:
                termFreq[word] = {}
            if x not in termFreq[word]:
                termFreq[word][x] = 0
            termFreq[word][x] += 1
    return DataFrame(termFreq).replace(to_replace = 'NaN', value=0).T

def idf(docs):
    """Find inverse document frequecny series from a dictionry of doucmnets."""
    wordOccurence = {}
    for x in docs:
        words = set(filter(lambda c: c not in string.punctuation, docs[x].replace('-', ' ')).lower().split())
        #words = set(docs[x].translate(None, punctuation).lower().replace("-", " ").split())
        for word in words:
            if word not in wordOccurence:
                wordOccurence[word] = 0
            wordOccurence[word] += 1

    idf = {}
    n = len(docs)
    for word in wordOccurence:
        idf[word] = log(float(n)/(1+wordOccurence[word]))
    return idf

def tf_idf(docs):
    """Return the product of the term-frequency and inverse document freqeucny."""
    tf = tfs(docs).T
    idfs = idf(docs)
    
    ordIdf = []
    for word in tf.columns.values:
        ordIdf.append(idfs[word])

    return tf.rmul(ordIdf, axis='columns').fillna(value=0).T
````

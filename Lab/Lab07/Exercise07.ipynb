### Estimating parameters of a logistic model

Gelman's book has an example where the dose of a drug may be affected to the number of rat deaths in an experiment.

| Dose (log g/ml) | # Rats | # Deaths |
|-----------------|--------|----------|
| -0.896          | 5      | 0        |
| -0.296          | 5      | 1        |
| -0.053          | 5      | 3        |
| 0.727           | 5      | 5        |

We will model the number of deaths as a random sample from a binomial distribution, where $n$ is the number of rats and $p$ the probabbility of a rat dying. We are given $n = 5$, but we believve that $p$ may be related to the drug dose $x$. As $x$ increases the number of rats dying seems to increase, and since $p$ is a probability, we use the following model:

$$
y \sim \text{Bin}(n, p) \\
\text{logit}(p) = \alpha + \beta x \\
\alpha \sim \mathcal{N}(0, 5) \\
\beta \sim \mathcal{N}(0, 10)
$$

where we set vague priors for $\alpha$ and $\beta$, the parameters for the logistic model.

```python
log_reg_code = """
data {
    int<lower=0> n;
    int rats;
    real x[n];
    int y[n];
}
transformed data {}
parameters {
    real a;
    real b;
}
transformed parameters {}
model {
    a ~ normal(0, 5);
    b ~ normal(0, 10);
    for(i in 1:n) {
        y[i] ~ binomial_logit(rats, b*x[i] + a);
  }
}
generated quantities {}
"""
# observed data
n = 4
rats = 5
x = np.array([-0.896, -0.296, -0.053, 0.727])
y = np.array([0, 1, 3, 5])

log_reg_dat = {
             'n': n,
             'x': x,
             'y': y,
             'rats': rats
            }

fit = pystan.stan(model_code=log_reg_code, data=log_reg_dat, iter=1000, chains=1)
```

```python
print fit
fit.plot(['a', 'b']);
plt.tight_layout()
```

### Using a hierarchcical model

This uses the Gelman radon data set and is based off this [IPython notebook](http://twiecki.github.io/blog/2014/03/17/bayesian-glms-3/). Radon levels were measured in houses from all counties in several states. Here we want to know if the preence of a basement affects the level of radon, and if this is affected by which county the house is located in. 

The data set provided is just for the state of Minnesota, which has 85 counties with 2 to 116 measurements per county. We only need 3 columns for this example `county`, `log_radon`, `floor`, where `floor=0` indicates that there is a basement.

We will perfrom simple linear regression on log_radon as a function of county and floor.


```python
radon = pd.read_csv('radon.csv')[['county', 'floor', 'log_radon']]
radon.head()
```

```python
hier_code = """
data {
    int<lower=0> n;
    int n_county;
    int county[n];
    real log_radon[n];
    int floors[n];
}
transformed data {}
parameters {
    real a[n_county];
    real b[n_county];
    real mu_a;
    real mu_b;
    real sigma;
    real sigma_a;
    real sigma_b;
}
transformed parameters {}
model {
    mu_a ~ normal(0,10000);
    mu_b ~ normal(0,10000);
    sigma_a ~ uniform(0,100);
    sigma_b ~ uniform(0,100);
    for (j in 1:n_county){
        a[j] ~ normal(mu_a, sigma_a);
        b[j] ~ normal(mu_b, sigma_b);
    }
    sigma ~ gamma(10,1);
    for(i in 1:n) {
        log_radon[i] ~ normal(a[county[i]] + b[county[i]]*floors[i], sigma);
  }
}
generated quantities {}
"""
# observed data
county = pd.Categorical(radon['county']).labels+1
n_county = len(set(county))
n = np.shape(radon)[0]
floors = radon.floor.values.astype("int")
log_radon = radon.log_radon.values
hier_dat = {
             'county': county,
             'n_county': n_county,
             'n': n,
             'floors': floors,
             'log_radon': log_radon
            }

fit = pystan.stan(model_code=hier_code, data=hier_dat, iter=1000, chains=1)
```

```python
print fit
fit.plot(['sigma_a', 'mu_a', 'sigma_b','mu_b']);
plt.tight_layout()
```

**Exercise 3 (10 points)**: Implement the leap-frog method and simulate the spring dynamics following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the spring trajectory plots.

```python
delta = 0.05 #step size

L = 130 # number of runs

def K(p):
    """ Kinetic energy function"""
    return (p**2)/2

def U(x):
    """Potential energy function"""
    return (x**2)/2

#Initial position
x0 = 1
p0 = 1

P=np.zeros(L+2)
X=np.zeros(L+1)

#Initial half step in momentum
P[0] = p0 - (delta/2)*x0 #dU(x)/dx = x

#Initial full step in x 
X[0] = x0 + delta*P[0]
 
for i in range(1,L+1):
    P[i] = P[i-1] - delta*X[i-1]
    X[i] = X[i-1] + delta*P[i]

#Final half step in momentum    
P[L+1] = P[L] - (delta/2)*X[L]
#
P=P[1:]

fig = plt.figure()
plt.plot(X,P)
fig.suptitle('Phase space')
plt.xlabel("x")
plt.ylabel("p")
```

**Exercise 4 (20 points)**: Now implement the HMC following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the bivariate normal MCMC plots. (For another simple explanation of HMC, see Chapter 14 of "Doing Basyesian Data Analysis" by Jouh Kruschke.) 

```python
np.random.seed(1000)

#Initial values are taken from the example provided
delta = 0.3
N = 1000 # number of iterations
L = 20 # number of leapfrog steps per iteration
 
def U(x):
    """Potential energy function"""
    sig=np.array([1,0.8,0.8,1]).reshape(2,2)  #Covariance matrix values taken from example
    return np.dot(np.dot(x,np.linalg.inv(sig)),x.T)
    
def dU(x):
    """Gradient of potential energy"""
    sig=np.array([1,0.8,0.8,1]).reshape(2,2)
    return np.dot(x,np.linalg.inv(sig))

def K(p):
    """Kinetic energy function"""
    return np.sum(np.dot(p.T,p))/2

#Initial state
x0=np.array([0,6]).reshape(2,1)
#Array to store states
X=np.zeros([2,N])
X[:,0]=x0.T


for t in range(1,N-1):
    #Sample a random p
    p0 = np.random.normal(0,1,2)
    
    ##first half step in p
    pStar = p0 - delta/2*dU(X[:,t-1])
    #first full step in x
    xStar = X[:,t-1] + delta*pStar     
    #L full steps
    for l in range(1,L):
        pStar = pStar - delta*dU(xStar)
        xStar = xStar + delta*pStar
     
    #final half step in p
    pStar = pStar - delta/2*dU(xStar)
    
    #Energies at the start and end of trajectory
    U0 = U(X[:,t-1])
    UStar = U(xStar)
 
    K0 = K(p0)
    KStar = K(pStar)
 
    #Acceptance probability
    alpha = min(1,np.exp((U0 + K0) - (UStar + KStar)))
 
    u = np.random.uniform(0,1)
    if u < alpha:
        X[:,t] = xStar
    else:
        X[:,t] = X[:,t-1]

fig = plt.figure()
plt.plot(X[0,:],X[1,:],'bo')
plt.plot(X[0,0:50],X[1,0:50])
fig.suptitle('HMC for bivariate normal')
plt.xlabel("x")
plt.ylabel("p")
plt.legend(('samples', 'First 50 states'), loc='best')
```
